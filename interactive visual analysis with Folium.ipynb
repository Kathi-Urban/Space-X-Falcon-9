 To answer your questions about the proximity of launch sites to the Equator and the coast, we can analyze the geographical locations shown in the image and their coordinates.

1. Proximity to the Equator
The Equator is at 0 degrees latitude. Launch sites closer to the Equator benefit from the rotational speed of the Earth, which can provide an additional velocity boost to rockets launched eastward.

From the image, we can observe the following launch sites:

Sites in Florida (e.g., Cape Canaveral, Kennedy Space Center) are relatively close to the Equator, at around 28 degrees latitude.
Sites on the West Coast (e.g., Vandenberg Space Force Base in California) are further from the Equator, at around 34 degrees latitude.
Sites in Texas (e.g., Boca Chica) are also relatively close to the Equator, at around 26 degrees latitude.
Findings: Not all launch sites are in close proximity to the Equator. Sites like those in California are significantly further from the Equator compared to those in Florida and Texas.

2.   Proximity to the Coast
Proximity to the coast is important for launch sites due to safety reasons (to avoid debris falling over populated areas) and ease of access for transportation of large rockets.

From the image:

All the launch sites marked are close to the coast.
The Florida sites are on the east coast.
The Texas site is near the Gulf of Mexico.
The California site is near the Pacific Ocean.
Findings: All launch sites shown in the image are in very close proximity to the coast, which is consistent with the safety and logistical considerations for launching rockets.

Explanation
Equator Proximity: While it is advantageous to be close to the Equator for the additional rotational speed, it is not a strict requirement. Launch sites can still be effective at higher latitudes, though they might require more fuel to achieve the same orbital velocities.
Coastal Proximity: Being near the coast is almost a necessity due to safety concerns, as launches over populated areas are avoided. Coastal sites allow rockets to launch over open water, minimizing risk to human populations.
In summary, while not all launch sites are near the Equator, they all tend to be near the coast, aligning with safety and logistical needs.

# Task 2: Mark the success/failed launches for each site on the map


Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.
Recall that data frame spacex_df has detailed launch records, and the `class` column indicates if this launch was successful or not


```python
spacex_df.tail(10)
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `launch_sites` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
def get_marker_color(cls): 
    return 'green' if cls == 1 else 'red' 

spacex_df['marker_color'] = spacex_df['class'].apply(get_marker_color) 
spacex_df.tail(10)
```

```python


# Function to determine the marker color based on launch success
def get_marker_color(launch_success):
    return 'green' if launch_success == 1 else 'red'

# Create a base map
site_map = folium.Map(location=[29.559684888503615, -95.0830971930759], zoom_start=5)

# Create a marker cluster
marker_cluster = MarkerCluster().add_to(site_map)

# Add markers to the map
for index, record in spacex_df.iterrows():
    coordinate = [record["Lat"], record["Long"]]
    name = record["Launch Site"]
    success = record["class"]
    marker_color = get_marker_color(success)
    
    # Create and add a Marker to the marker cluster
    marker = folium.Marker(
        location=coordinate,
        icon=folium.Icon(color=marker_color),
        popup=name
    )
    marker_cluster.add_child(marker)
    
    # Add a custom HTML marker
    folium.Marker(
        location=coordinate,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{name}</b></div>'
        )
    ).add_to(site_map)

# Display the map
site_map
```

# TASK 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
